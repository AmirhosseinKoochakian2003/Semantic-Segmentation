# Packages

```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, MaxPooling2D, Dropout, concatenate, UpSampling2D, ZeroPadding2D, Add, Cropping2D
from tensorflow.keras import Model
from tensorflow.keras.optimizers import Adam

import PIL.Image, PIL.ImageFont, PIL.ImageDraw
```

# Dataset

```python
BATCH_SIZE = 32
```

```python
def load_images_split() :
    # loading the images and segmentation masks
    images = np.load('/kaggle/input/multimnistm2nist/combined.npy')
    segments = np.load('/kaggle/input/multimnistm2nist/segmented.npy')
    
    # split the data to train, val, test
    train_images, test_images, train_masks, test_masks = train_test_split(images, segments, test_size=0.2, shuffle=True, random_state=42)
    val_images, test_images, val_masks, test_masks = train_test_split(test_images, test_masks, test_size=0.5, shuffle=True, random_state=42)
    
    return (train_images, train_masks), (val_images, val_masks), (test_images, test_masks)

def preprocess(image, mask) :
    image = tf.cast(image, dtype=tf.float32)
    mask = tf.cast(mask, dtype=tf.float32)
    
    # gray scale
    image = tf.reshape(image, (image.shape[0], image.shape[1], 1))

    # Min-Max normalization
    image = image / 127.5
    image -= 1
    
    return image, mask

def load_trainset(images, masks):
    trainset = tf.data.Dataset.from_tensor_slices((images, masks))
    trainset = trainset.map(preprocess)

    trainset = trainset.shuffle(5000, reshuffle_each_iteration=False)
    trainset = trainset.batch(BATCH_SIZE)
    trainset = trainset.prefetch(-1)

    return trainset

def load_valset(images, masks) :
    valset = tf.data.Dataset.from_tensor_slices((images, masks))
    valset = valset.map(preprocess)

    valset = valset.batch(BATCH_SIZE)

    return valset

def load_testset(images, masks) :
    testset = tf.data.Dataset.from_tensor_slices((images, masks))
    testset = testset.map(preprocess)

    testset = testset.batch(BATCH_SIZE)

    return testset   
```

```python
train, val, test = load_images_split()
```

```python
train_dataset = load_trainset(train[0], train[1])
val_dataset = load_valset(val[0], val[1])
test_dataset = load_testset(test[0], test[1])
```

```python
train_iterator = train_dataset.as_numpy_iterator().next()
width = train_iterator[0].shape[1]
height = train_iterator[0].shape[2]
```

# Visualizing

```python
# there are 11 classes in the dataset: one class for each digit (0 to 9) plus the background class
n_classes = 11

# assign a random color for each class
colors = [tuple(np.random.randint(256, size=3) / 255.0) for i in range(n_classes)]
```

## Using PIL

```python
def fuse_with_pil(images):
    widths = (image.shape[1] for image in images)
    heights = (image.shape[0] for image in images)
    total_width = sum(widths)
    max_height = max(heights)

    new_im = PIL.Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in images:
        pil_image = PIL.Image.fromarray(np.uint8(im))
        new_im.paste(pil_image, (x_offset,0))
        x_offset += im.shape[1]
  
    return new_im


def give_color_to_annotation(annotation):
    seg_img = np.zeros( (annotation.shape[0],annotation.shape[1], 3) ).astype('float')
  
    for c in range(n_classes):
        segc = (annotation == c)
        seg_img[:,:,0] += segc*( colors[c][0] * 255.0)
        seg_img[:,:,1] += segc*( colors[c][1] * 255.0)
        seg_img[:,:,2] += segc*( colors[c][2] * 255.0)
  
    return seg_img


def show_annotation_and_prediction(image, annotation, prediction):
    new_ann = np.argmax(annotation, axis=2)
    true_img = give_color_to_annotation(new_ann)
    pred_img = give_color_to_annotation(prediction)

    image = image + 1
    image = image * 127.5
    image = np.reshape(image, (image.shape[0], image.shape[1],))
    image = np.uint8(image)
    images = [image, np.uint8(pred_img), np.uint8(true_img)]

    

def show_annotation_and_image(image, annotation):
    new_ann = np.argmax(annotation, axis=2)
    seg_img = give_color_to_annotation(new_ann)

    image = image + 1
    image = image * 127.5
    image = np.reshape(image, (image.shape[0], image.shape[1],))

    image = np.uint8(image)
    images = [image, seg_img]

    images = [image, seg_img]
    fused_img = fuse_with_pil(images)
    plt.imshow(fused_img)


def list_show_annotation(dataset, num_images):
    ds = dataset.unbatch()

    plt.figure(figsize=(20, 15))
    plt.title("Images And Annotations")
    plt.subplots_adjust(bottom=0.1, top=0.9, hspace=0.05)

    for idx, (image, annotation) in enumerate(ds.take(num_images)):
        plt.subplot(5, 5, idx + 1)
        plt.yticks([])
        plt.xticks([])
        show_annotation_and_image(image.numpy(), annotation.numpy())
```

```python
list_show_annotation(train_dataset, 10)
```

## Using PLT

```python
def create_mask(pred_mask):
    pred_mask = tf.argmax(pred_mask, axis=-1)
    pred_mask = tf.expand_dims(pred_mask, axis=-1)
    return pred_mask[0]
```

```python
def display(display_list, file_name):
    plt.figure(figsize=(12, 12))

    title = ['Input Image', 'True Mask', 'Predicted Mask']

    for i in range(len(display_list)):
        plt.subplot(1, len(display_list), i+1)
        plt.title(title[i])
        plt.imshow(tf.keras.preprocessing.image.array_to_img(display_list[i]))
        plt.axis('off')
    plt.savefig(f'{file_name}.jpg', dpi=300)
    plt.show()
```

```python
def give_color_to_annotation_PLT(annotation, colors):
    seg_img = tf.zeros((annotation.shape[0], annotation.shape[1], 3), dtype=tf.float32)

    for c in range(len(colors)):
        segc = tf.cast(annotation == c, dtype=tf.float32)  # Convert to float
        color_tensor = tf.cast(tf.constant(colors[c]), dtype=tf.float32)
        seg_img = tf.math.add(seg_img, segc * color_tensor * 255.0)

    return seg_img
```

```python
def visualize_predictions(model, dataset, num, type_):
    dataset_unbatched = dataset.unbatch()
    dataset_unbatched = dataset_unbatched.batch(1)
    
    i = 0

    for image, mask in dataset_unbatched :
        if i == num :
            break
        pred_mask = model.predict(image)
        true_mask_colored = give_color_to_annotation_PLT(create_mask(mask), colors)
        pred_mask_colored = give_color_to_annotation_PLT(create_mask(pred_mask), colors)
        display([image[0], true_mask_colored, pred_mask_colored], f'{type_}{i}')
        i += 1
```

# Metrics : IoU & Dice score

```python
def calc_metrics_image_wise(y_true, y_pred) :
    iou_list = []
    dice_list = []
    for i in range(y_true.shape[0]) : 
        y = y_true[i]
        pred = y_pred[i]
        
        image_iou = {}
        image_dice = {}

        for j in range(n_classes):
            overlap_area = np.sum((pred == j) * (y == j))
            y_pred_area = np.sum((pred == j))
            y_true_area = np.sum((y == j))
            union_area = y_true_area + y_pred_area
            
            image_iou[j] = (overlap_area) / (union_area - overlap_area + 0.000001)
            image_dice[j] = 2 * (overlap_area) / (union_area + 0.000001)

        iou_list.append(image_iou)
        dice_list.append(image_dice)

    return iou_list, dice_list
```

```python
def calc_metrics_class_wise(y_true, y_pred) :
    iou_list = []
    dice_list = []

    for i in range(n_classes):
        iou = 0
        dice = 0
        count = 0
        for j in range(y_true.shape[0]):
            y = y_true[j]
            pred = y_pred[j]
            
            overlap_area = np.sum((pred == i) * (y == i))
            y_pred_area = np.sum((pred == i))
            y_true_area = np.sum((y == i))
            union_area = y_true_area + y_pred_area
            
            if y_true_area > 0:
                iou += (overlap_area) / (union_area - overlap_area + 0.000001)
                dice += 2 * (overlap_area) / (union_area + 0.000001)
                count += 1
                
        iou_list.append(iou / count + 0.000001)
        dice_list.append(dice / count + 0.000001)
    return iou_list, dice_list
```

```python
def call_metric_function(dataset, model) :
    y_true = []
    iterator = dataset.as_numpy_iterator()
    batch = next(iterator, None)
    
    while batch != None : 
        for mask in batch[1] :
            y_true.append(np.argmax(mask, axis=2))
        batch = next(iterator, None) 
    y_true = np.array(y_true)

    y_pred = model.predict(dataset)
    y_pred = np.argmax(y_pred, axis=3)
    
    metrics_images = calc_metrics_image_wise(y_true, y_pred)
    metrics_classes = calc_metrics_class_wise(y_true, y_pred)
    
    return metrics_images[0], metrics_images[1], metrics_classes[0], metrics_classes[1]
```

```python
def print_result(iou_image, dice_image, iou_class, dice_class, model_type, data_type, idx):
    print(f'{model_type} {data_type} IoU :')
    for index, element in enumerate(iou_class):
        print(f'class : {index}, IoU : {element:.4f}')
    
    print('\n====================\n')
        
    print(f'{model_type} {data_type} Dice score :')
    for index, element in enumerate(dice_class):
        print(f'class : {index}, Dice score : {element:.4f}')
    
    print('\n====================\n')

    iou = iou_image[idx]
    dice = dice_image[idx]
    iou = {key: value for key, value in iou.items() if value > 0}
    dice = {key: value for key, value in dice.items() if value > 0}
    
    print(f'{model_type} {data_type} IoU on image {idx} :')
    for key, value in iou.items():
        print(f'class : {key}, IoU : {value:.4f}')

    print('\n====================\n')
    
    print(f'{model_type} {data_type} Dice score on image {idx} :')
    for key, value in dice.items():
        print(f'class : {key}, Dice score : {value:.4f}')
```

# U-Net

## Hyperparameters

```python
EPOCHS = 10
```

## Architecture

```python
def downsampling_block(X=None, number_of_filters=32, dropout_rate=0, max_pooling=True) :
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    
    if dropout_rate > 0 :
        X = Dropout(dropout_rate)(X)
        
    down_activation = X
    
    if max_pooling :
        down_activation = MaxPooling2D(pool_size=(2, 2))(X)
        
    skip_connection_activation = X
    
    return down_activation, skip_connection_activation
```

```python
def upsampling_block(prev_layer_activation=None, skip_connection_activation=None, number_of_filters=32) :
    up_activation = UpSampling2D(size=(2, 2))(prev_layer_activation)
    
    # Check if widths are not equal and adjust if necessary
    if prev_layer_activation.shape[2] != skip_connection_activation.shape[2]:
        target_width = skip_connection_activation.shape[2]
        pad_width = max(target_width - up_activation.shape[2], 0)
        padding = ((0, 0), (0, pad_width))
        up_activation = ZeroPadding2D(padding=padding)(up_activation)

    X = concatenate([up_activation, skip_connection_activation], axis=3)
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    
    return X
```

```python
def Unet_model(input_shape=(64, 84, 1), number_of_filters=32, number_of_classes=11) :
    inputs = Input(input_shape)
    
    down1 = downsampling_block(inputs, number_of_filters)
    down2 = downsampling_block(down1[0], number_of_filters * 2)
    down3 = downsampling_block(down2[0], number_of_filters * 4)
    down4 = downsampling_block(down3[0], number_of_filters * 8, dropout_rate=0.2)
    down5 = downsampling_block(down4[0], number_of_filters * 16, dropout_rate=0.2, max_pooling=False)
    
    up1 = upsampling_block(down5[0], down4[1], number_of_filters * 8)
    up2 = upsampling_block(up1, down3[1], number_of_filters * 4)
    up3 = upsampling_block(up2, down2[1], number_of_filters * 2)
    up4 = upsampling_block(up3, down1[1], number_of_filters)
    
    
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(up4)
    Y = Conv2D(number_of_classes, 1, activation='softmax', padding='same')(X)
    
    model = Model(inputs=inputs, outputs=Y)
    
    return model
```

```python
unet_model = Unet_model((width, height, 1), 32, 11)
```

```python
unet_model.summary()
```

## Loss function & Optimizer

```python
unet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

## Training

# 2D V-Net

## Hyperparameters

```python
EPOCHS = 30
```

## Architecture

```python
def downsampling_conv_block(X=None, number_of_filters=32, dropout_rate=0, downsample=True) :
    residual = X
    residual = Conv2D(number_of_filters, 1, activation='linear', padding='same', kernel_initializer='he_normal')(residual) 
    
    A = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    A = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(A)
    A = Add()([residual, A])

    if dropout_rate > 0 :
        A = Dropout(dropout_rate)(A)
        
    down_activation = A
    skip_connection_activation = A
    
    if downsample:
        down_activation = Conv2D(number_of_filters, 2, activation='relu', strides=(2, 2), padding='valid', kernel_initializer='he_normal')(A)
    
    return down_activation, skip_connection_activation
```

```python
def upsampling_conv_block(prev_layer_activation=None, skip_connection_activation=None, number_of_filters=32) :
    up_activation = Conv2DTranspose(number_of_filters, 3, strides=(2, 2), padding='same')(prev_layer_activation)
    
    residual = up_activation
    
    X = concatenate([up_activation, skip_connection_activation], axis=3)
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(X)
    X = Add()([residual, X])

    return X
```

```python
def Vnet_model(input_shape=(64, 84, 1), number_of_filters=32, number_of_classes=11):
    inputs = Input(input_shape)
    # pad the input image height to have dimensions to the nearest power of two
    padded_inputs = tf.keras.layers.ZeroPadding2D(padding=(0, 6))(inputs)
    
    down1 = downsampling_conv_block(padded_inputs, number_of_filters)
    down2 = downsampling_conv_block(down1[0], number_of_filters * 2)
    down3 = downsampling_conv_block(down2[0], number_of_filters * 4)
    down4 = downsampling_conv_block(down3[0], number_of_filters * 8, dropout_rate=0.3)
    down5 = downsampling_conv_block(down4[0], number_of_filters * 16, dropout_rate=0.3, downsample=False)
    
    up1 = upsampling_conv_block(down5[0], down4[1], number_of_filters * 8)
    up2 = upsampling_conv_block(up1, down3[1], number_of_filters * 4)
    up3 = upsampling_conv_block(up2, down2[1], number_of_filters * 2)
    up4 = upsampling_conv_block(up3, down1[1], number_of_filters)
    
    X = Conv2D(number_of_filters, 3, activation='relu', padding='same', kernel_initializer='he_normal')(up4)
    Y = Conv2D(number_of_classes, 1, activation='softmax', padding='same')(X)
    Y = Cropping2D(((0, 0), (0, 12)))(Y)
    
    model = Model(inputs=inputs, outputs=Y)
    
    return model
```

```python
vnet_model = Vnet_model((width, height, 1), 32, 11)
```
